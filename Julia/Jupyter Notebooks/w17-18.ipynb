### Tasks
- neuron distribution w.r.t. brain region for each mouse
- plot correlation matrix, log mutual information matrix, distance matrix and brain area difference matrix for each mouse
- plot spike train matrix with neurons rearranged by brain area

```julia
using Plots, Measures; gr()
using MAT, Printf
using LinearAlgebra, Statistics, Random
using Clustering, Distances
using InformationMeasures
using Ripserer
using BenchmarkTools
```

```julia
# Load all data in a three-element array of dictionaries
# Each element corresponding to each mouse maps variable names to their contents  
mice_names = ["Krebs", "Waksman", "Robbins"]
data = [MAT.matread("../Data/source/$(mice_names[i])withFaces_KS2.mat") for i in 1:3];
```

### Brain Regions Disttribution

```julia
function dataDist(dt, names, dttype)
    # neuron distribution w.r.t. height on probe
    if dttype == :height
        p1 = histogram(dt[1]["Wh"], nbin=:rice, title=names[1])
        p2 = histogram(dt[2]["Wh"], nbin=:rice, title=names[2])
        p3 = histogram(dt[3]["Wh"], nbin=:rice, title=names[3])
        plot(p1, p2, p3, xlabel="Height on probe", yaxis=("Number of neurons", (0, 300)),
            guidefontsize=10, layout=(1, 3), size=(900, 400), grid=false, legend=false)
        
    # neuron distribution w.r.t. brain area    
    elseif dttype == :area
        p1 = histogram(dt[1]["brainLoc"], nbin=1:15, title=names[1])
        p2 = histogram(dt[2]["brainLoc"], nbin=1:15, title=names[2])
        p3 = histogram(dt[3]["brainLoc"], nbin=1:15, title=names[3])
        plot(p1, p2, p3, xticks=(1:14, vec(data[1]["areaLabels"])), xrotation=90, xlabel="Brain region", yaxis=("Number of neurons", (0, 2000)),
            guidefontsize=10, layout=(1, 3), size=(900, 400), grid=false, legend=false, left_margin=2mm, bottommargin=10mm)
    
    # neuron distribution w.r.t. probes
    elseif dttype == :probe
        p1 = histogram(dt[1]["iprobe"], nbin=8, title=names[1])
        p2 = histogram(dt[2]["iprobe"], nbin=8, title=names[2])
        p3 = histogram(dt[3]["iprobe"], nbin=8, title=names[3])
        plot(p1, p2, p3, xlabel="Probe", yaxis=("Number of neurons", (0, 700)),
            guidefontsize=10, layout=(1, 3), size=(900, 400), grid=false, legend=false)
    
    # neuron distribution w.r.t. firing rates
    elseif dttype == :rate
        p1 = histogram(sum(dt[1]["stall"], dims=2)/(dt[1]["tspont"][end]-dt[1]["tspont"][1]), nbin=18, title=names[1])
        p2 = histogram(sum(dt[2]["stall"], dims=2)/(dt[2]["tspont"][end]-dt[2]["tspont"][1]), nbin=18, title=names[2])
        p3 = histogram(sum(dt[3]["stall"], dims=2)/(dt[3]["tspont"][end]-dt[3]["tspont"][1]), nbin=18, title=names[3])
        plot(p1, p2, p3, xaxis=("Firing Rate", (0, 100)), yaxis=("Number of neurons", (0, 1000)),
            guidefontsize=10, layout=(1, 3), size=(900, 400), grid=false, legend=false)
    end
end
```

```julia
dataDist(data, mice_names, :area)
```

### "All-neuron" heatmaps

```julia
function corClust(cmat, mtr, lnk)
    sym2func = Dict(:CosD => Distances.CosineDist(), :EucD => Distances.Euclidean())
    
    if mtr == :None
        sort_idx = 1:size(cmat, 1)
    else
        diss_mat = Distances.pairwise(sym2func[mtr], cmat, dims=1)
        result = Clustering.hclust(diss_mat, linkage=lnk)
        sort_idx = result.order
    end
    
    return sort_idx
end


function diff(v, binary=false)
    rep_mat = repeat(v, 1, length(v))
    diff_mat = abs.(rep_mat - transpose(rep_mat))
    
    if binary
        diff_mat = diff_mat .!= 0
    end
    
    return diff_mat
end

function mutualInfo(st, filter=nothing, fileIO=:none, id=0)
    if fileIO == :read
        file = MAT.matopen("../Data/save/M$(id).mat")
        output = MAT.read(file, "logmi")
        if filter != nothing
            output = output[filter, filter]
        end
        MAT.close(file)
    else
        if filter != nothing
            st = st[filter, :]
        end
        n = size(st, 1)
        mi_mat = zeros(n, n)

        for i=1:n-1
            for j=i+1:n
                mi_mat[i, j] = InformationMeasures.get_mutual_information(st[i, :], st[j, :])
            end
        end

        mi_mat = mi_mat + transpose(mi_mat)

        for i=1:n
            mi_mat[i, i] = InformationMeasures.get_entropy(st[i, :])
        end
        
        output = log.(mi_mat)
        
        if fileIO == :write && filter == nothing
            file = MAT.matopen("../Data/save/M$(id).mat", "w")
            MAT.write(file, "logmi", output)
            MAT.close(file)
        end
    end
    
    return output
end

function clustWithinRegion(D, reg, regsort)
    D = D[regsort, regsort]
    reg = reg[regsort]
    
    for i=1:14
        filter = reg .== i
        if sum(filter) != 0
            res = Clustering.hclust(D[filter, filter], linkage=:single)
            regsort[filter] = regsort[filter][res.order]
        end
    end
    
    return regsort
end
    
function spkCorCompare(data, id, link=:none, metric=:none, byprobe=false)
    spktrain = data[id]["stall"]          # multi-neuron spike train matrix
    hop = data[id]["Wh"]                  # heights of neurons on probes
    coord = data[id]["ccfCoords"]         # CCF coordinates of neurons
    regIDs = data[id]["brainLoc"]         # index labels of brain areas
     
    if byprobe
        probeIDs = data[id]["iprobe"]      # IDs of corresponding probes
        plot_array = Array{Plots.Plot{Plots.GRBackend}}(undef, 8*3)
        
        for i=1:8
            extract_idx = vec(probeIDs .== i)
            cor_mat_iprobe = Statistics.cor(spktrain[extract_idx, :], dims=2)
            if link == :none || metric == :none
                sort_idx = sortperm(hop[extract_idx])
            else
                sort_idx = corClust(cor_mat_iprobe, metric, link)
            end
            dist_mat_iprobe = diff(hop[extract_idx])
            logmi_mat_iprobe = mutualInfo(spktrain, extract_idx, :read, id)
            
            plot_array[i*3-1] = heatmap(cor_mat_iprobe[sort_idx, sort_idx], clim=(-0.5, 1), title="M$id-P$i-cor-$metric")         
            plot_array[i*3-2] = heatmap(dist_mat_iprobe[sort_idx, sort_idx], clim=(0, 4000), title="M$id-P$i-dist-$metric")
            plot_array[i*3] = heatmap(logmi_mat_iprobe[sort_idx, sort_idx], clim=(-20, 2.5), title="M$id-P$i-logmi-$metric")
        end
        
        display(plot(plot_array[1:12]..., layout=(4,3), size=(1500, 1500), yflip=true))
        #savefig("../Plots/M$(id)_P1-4_$(metric)-$(link).png")
        display(plot(plot_array[13:24]..., layout=(4,3), size=(1500, 1500), yflip=true))
        #savefig("../Plots/M$(id)_P5-8_$(metric)-$(link).png")
        
    else
        cor_mat_all = Statistics.cor(spktrain, dims=2)
        dist_mat_all = Distances.pairwise(Distances.Euclidean(), coord, dims=1)
        if link == :none || metric == :none
            sort_idx = sortperm(vec(regIDs))
            #sort_idx = clustWithinRegion(dist_mat_all, regIDs, sort_idx)
        else
            sort_idx = corClust(cor_mat_all, metric, link)
        end
        logmi_mat_all = mutualInfo(spktrain, nothing, :read, id)
        reg_mat_all = diff(regIDs, true)
        
        p1 = heatmap(cor_mat_all[sort_idx, sort_idx], clim=(-0.3, 1), title="M$id-all-cor-$metric-$link")
        p2 = heatmap(logmi_mat_all[sort_idx, sort_idx], clim=(-20, 1.1), title="M$id-all-logmi-$metric-$link")
        p3 = heatmap(dist_mat_all[sort_idx, sort_idx], clim=(0, 6050), title="M$id-all-dist-$metric-$link")
        p4 = heatmap(reg_mat_all[sort_idx, sort_idx], color=cgrad(:default, 2, categorical=true), title="M$id-all-reg-$metric-$link")
        plot(p1, p2, p3, p4, layout=(2, 2), size=(1450, 1200), yflip=true)
        savefig("../Plots/M$(id)_all_$(metric)-$(link).png")
    end
end
```

### Spike Train Heatmaps

```julia
function spktrainDisp(data, id, binary=false)
    spktrain = data[id]["stall"]
    regIDs = data[id]["brainLoc"]
    reglbs = data[id]["areaLabels"]
    
    cap = binary ? 1 : 36      # 36 is the highest number of spikes for a single sample and single neuron
    N, T = size(spktrain)      # N : number of neurons, T : number of samples (sampling rate x recording time)

    unique_regIDs = convert(Array{Int64, 1}, sort(unique(regIDs)))
    unique_regIDs = unique_regIDs[1] == 0 ? unique_regIDs[2:end] : unique_regIDs
    n_reg = length(unique_regIDs)
    plot_holder = Array{Plots.Plot{Plots.GRBackend}}(undef, n_reg)
    plot_heights = Array{Float64, 1}(undef, n_reg)

    for i=1:n_reg
        filter = vec(regIDs .== unique_regIDs[i])
        plot_heights[i] = round(sum(filter)/N; digits=3)
        plot_holder[i] = heatmap(spktrain[filter, :], title=reglbs[unique_regIDs[i]],
                         color=cgrad(:grays, rev=true), axis=false, ticks=false)
    end
    
    plot(plot_holder..., clims=(0, cap), yticks=0:50:1000, title_position=:left, left_margin=0mm,
         layout=grid(n_reg, 1, heights=plot_heights), size=(round(T/10), N), colorbar=false)
    savefig("../Plots/spktrain_M$(id).png")
    
#     heatmap(spktrain[sort_idx, :], clim=(0, cap), color=cgrad(:grays, rev=true), yflip=true,
#             size=(round(T/10), round(N/2)), colorbar=false)
    #savefig("../Plots/spktrain_M$(id).png")
end
```
